```python
#Country Filter
def Country(soup):
    country = []
    reg = re.compile('.*country_of_origin')
    try:
        div = soup.find('div',{'id':'titleDetails'})
        a2 = div.find_all('a',{'href':reg})
        for a in a2:
            country.append(a.text)
    except:
        country = 'None'
    return country

import re
import requests
from bs4 import BeautifulSoup
country = []
for i in range(0,len(index)):
    try:
        string = 'https://www.imdb.com/title/{}/'.format(index[i])
        result = requests.get(string)
        src = result.content
        soup = BeautifulSoup(src, 'html.parser')
        country.append(Country(soup))
    except:
        country.append('None')
```

```python
def Rating(soup):
    rate = []
    try:
        div = soup.find('div',{'class':'imdbRating'})
        span = div.find('span',{'itemprop':'ratingValue'})
        rate = span.text
    except:
        rate = '0'
    return rate
```

```python
def Direct(soup):
    direct = []
    path = re.compile('^/name/nm.*')
    director = re.compile('(Director:|Directors:)')
    form = re.compile('^(?!\n$).*')
    try:
        div = soup.find('div',{'class':'plot_summary'})
        h4 = div.find('h4',text=director)
        a = h4.find_next_sibling('a',{'href':path})
        direct = a.text
    except:
        direct = 'None'
    return direct
```

```python
def Cast(soup):
    cast = []
    path = re.compile('^/name/nm.*')
    star = re.compile('(Stars:|Star:)')
    try:
        div = soup.find('div',{'class':'plot_summary'})
        h4 = div.find('h4',text=star)
        a2 = h4.find_next_siblings('a',{'href':path})
        for a in a2:
            cast.append(a.text)
    except:
        cast = 'None'
    return cast
```

```python
def Genre(soup):
    genre = []
    path2 = re.compile('^/search/title\?genre.*')
    try:
        div = soup.find('div',{'class':'article','id':'titleStoryLine'})
        for a in div.find_all('a',{'href':path2}):
            genre.append(a.text)
    except:
        genre = 'None'
    return genre
```

```python
def Release(soup):
    try:
        h4 = soup.find('h4',text='Release Date:')
        g = h4.find_all_next(string=True)[1]
        reg = re.compile('\d+\s\w+\s\d+')
        release = re.findall(reg,g)[0]
    except:
        release = 'None'
    return(release)
release=[]
```

```python
def Budget(soup):
    try:
        h4 = soup.find('h4',text='Budget:')
        g = h4.find_all_next(string=True)[1]
        reg = re.compile('\$.*\d')
        budget = re.findall(reg,g)[0]
    except:
        budget = '0'
    return budget
```

```python
def Time(soup):
    try:
        h4 = soup.find('h4',text='Runtime:')
        time = h4.find_all_next(string=True)[2]
    except:
        time = 'None'
    return time
```
