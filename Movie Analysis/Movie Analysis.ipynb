```python
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(raw_df6,raw_df6['Year']):
    strat_train_set = raw_df6.iloc[train_index,:]
    strat_test_set = raw_df6.iloc[test_index,:]

Y_train = strat_train_set['y'].values
x_train = strat_train_set.iloc[:,1:-1].values
Y_test = strat_test_set['y'].values
x_test = strat_test_set.iloc[:,1:-1].values
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(random_state=42)

from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=100,
                                n_jobs=-1, random_state=42)

from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=30,
                         algorithm="SAMME.R",
                         learning_rate=0.25, random_state=42)

from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(max_depth=2,
                                n_estimators=5,
                                learning_rate=0.1,
                                random_state=42)
```

```python
from sklearn.ensemble import VotingClassifier
vote_s = VotingClassifier(estimators=[('forest', forest),
                                      ('tree', treeclf),
                                      ('ada', ada), ('gb', gb)],
                          n_jobs=-1,
                          voting='soft')
vote_s.fit(x_train, Y_train)
vote_s.score(x_test, Y_test)
```

```python
df = pd.read_csv("predition_data.csv")
x_t = df.iloc[:,1:-1].values
Y_t = df['y']
vote_s.score(x_t,Y_t)
```
